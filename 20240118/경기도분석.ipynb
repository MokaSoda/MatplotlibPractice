```python
import  pandas as pd

file = '경기도행정구역경계.json'

df = pd.read_json(file, orient='columns')
df
```

```python

# json normalize using pd
df_normalized = pd.json_normalize(df['features'], max_level=0)

df_normalized.drop(columns='type')
```

```python

df2 = pd.read_json('repo.json')
df2_owner = pd.json_normalize(df2['owner'])
df2.drop(columns=['owner'], inplace=True)

for x in df2_owner.columns:
    if x in df2.columns:
        df2_owner.drop(columns=x, inplace=True)

combinedDF = df2_owner.combine_first(df2).set_index('id')
combinedDF[['full_name', 'description']]
```
