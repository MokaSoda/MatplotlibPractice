## [데이터 시각화 맛보기]

(1) 모듈 로딩 <hr>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

(2) 데이터 로딩 <hr>

```python
# 2D 그래프 그리기 위한 데이터 준비
xx = range(30)
y = [x**2 for x in xx]

xx2 = range(30)
y2 = [x**1.8 for x in xx]

```

(3) 데이터 시각화 <hr>

- [기본] 라인/선 그래프 
    * 시간의 변화를 표현해주는 그래프
    * 연속적/순서 있는 데이터를 시각화할 경우 주로 사용


```python
# 데이터를 데이터프레임으로 변환 
# plot(data) ==> y축 데이터로 사용
# plot(data1, data2) ==> data1은  x축, data2는 y축
# 반환값 : 라인/선 그래프 객체 리스트에 담겨서 반환됨
```

- 라인/선 그래프에서 형식 즉 포맷지정 => str 타입의 포멧 설정
    * '색성형태선스타일' ==> 'ro--' : red circle dot-line
    * 대표적으로 많이 사용되는 색상: r - red, g - green, b - blue, k - black
    * 대표적으로 많이 사용하는 형태 :
        - o : circle
        - ^ : triangle
        - s : square
        - v : triangle_down
        - < : triangle_left
        - \> : triangle_right
        - 8 : octagon
    *  대표적으로 많이 사용되는 선 스타일: 
        * \- : solid line, \-\- : dashed line, (\:) : dotted line, -. : dash-dot line

```python
print('v' == 'v')
test = plt.plot(xx,y,'r8:',xx2, y2, 'b')
```

```python

# 리스트 안에 그래프가 저장되므로 리스트 인덱스로 접근 해야함
print(type(test), len(test), type(test[0]))

plt.show()
```
