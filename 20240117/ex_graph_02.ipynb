```python
# 그래프 관련 모듈
import matplotlib.pyplot as plt
# 패키지
import matplotlib as mpl
# 폰트 관련 모듈
import matplotlib.font_manager as fm

import pandas as pd
```

```python
# 2D 그래프 그리기 위한 데이터 준비
xx = range(30)
y = [x**2 for x in xx]

xx2 = range(30)
y2 = [10*x for x in xx]

```

- 한글 폰트 설정


```python
# 한글 폰트 설정
# 폰트 리스트 출력
# 직접 폰트 파일의 경로 지정하여 맑은 고딕 사용하기
font_path = 'C:/Windows/Fonts/NanumGothic.ttf'
font_prop = fm.FontProperties(fname=font_path).get_name()
mpl.rc('font', family=font_prop)

# get current font name
print(mpl.rcParams['font.family'])
```

```python
plt.title("제목")

print('v' == 'v')
plt.plot(xx,y,'r8:',xx2, y2, 'b')
plt.show()
plt.close()
```

### x축, y축 레이블 설정 및 눈금 설정
- x, y 축 데이터에 대한 설명 => plt.xlabel(), plt.ylabel()
- 눈금 간격, 출력 형태를 설정가능함 => plt.xtikes(), plt.yticks(), plt.tick_params()




```python
# 그래프 그리기 
plt.plot(xx,y,'r8:',xx2, y2, 'b')
# plt.title("제목 입력됨", color='yellow', fontsize=30)
# make title use different font family
# i want to input y=x^(1.8) as latex in plt title
plt.title(r"$y=x^2, y=10x$", fontsize=30, fontweight='bold')
plt.xlabel("input value", loc='right', color='g', fontsize=30, rotation=45)


# plt.ylabel("output value", loc='top')
# make ylabel font bigger
# plt.ylabel("output value", fontsize=15)
# make ylabel font direction to normal
plt.ylabel("output value", loc='top',fontsize=30, rotation=45)

# make plot resolution grid
# plt.grid(True)

# 눈금 간격 설정
labels=['a','b','c','d','e','f','g']
plt.xticks(ticks=range(0,31,5) ,color='green', fontsize = 30)
# make xticks font bigger

# make xticks color blue
plt.yticks(ticks=range(0,1001,50))
# make yticks use custom color code f21de3
plt.yticks(ticks=range(0,1001,50), color='#f21de3')

# 눈금 출력 형태 설정
plt.tick_params(axis='both', which='major', labelsize=15)
# make plt.legend under graph
# plt.legend([r'$y=x^2$', r'$y=10x$'], loc='lower right')
# make legend outside of chart using bbox_to_anchor 

#범래추가 ==> 여러개의 그래프를 표시하는 경우 그래프별 식별을 위해서 사용
plt.legend([r'$y=x^2$', r'$y=10x$'], loc='upper left',bbox_to_anchor=(0, 0.3))
# show one point of graph it's real number and label it's number

# 텍스트 추가 => 그래프 내에 위치 지정 후 텍스트 추가 설정 => plt.text()
# plt.text(0.5, 500, '텍스트추가', fontsize=30, color='#42ff12')
# fill textbox black color
box = {
'facecolor' : '.75',
'edgecolor' : 'k',
'boxstyle' : 'round4',
'alpha': 1,
}

plt.text(0.5, 500, '텍스트추가', fontsize=30, color='#42ff12',bbox=box)

# ### 화살표 설정
# - 그래프에 특정 지점에 화살표 표기
# - text와 함께 종종 사용됨
# - plt.annotate() 함수 사용

# i want to point to specific dot using arrow with plt.annotate
# plt.annotate('15,150 좌표', xy=(15, 150), xytext=(30, 100),
# arrowprops=dict(arrowstyle = '->'))
# i want to change '->' arrow color
plt.annotate('15,150 좌표', xy=(15, 150), xytext=(30, 100),
arrowprops=dict(arrowstyle = '->', color='#f21de3'))
```
